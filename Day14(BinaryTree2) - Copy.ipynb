```python
class Solution:
    
    def diameterOfTree(self, node: TreeNode) -> List:
        if not node:
            return [0, 0]
        
        diameterLeft, longestBranchLeft = self.diameterOfTree(node.left)
        diameterRight, longestBranchRight = self.diameterOfTree(node.right)
        currentDiameter = longestBranchLeft + longestBranchRight + 1
        
        return [max(diameterLeft, diameterRight, currentDiameter), max(longestBranchLeft, longestBranchRight) + 1]
    
    def diameterOfBinaryTree(self, root: TreeNode) -> int:
        return max(self.diameterOfTree(root)) - 1 if root else 0
```

# Level Order Traversal

```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def printTreeDetailed(root):
    if root==None:
        return 0
    q=queue.Queue()
    q.put(root)
    q.put(None)
    while (not(q.empty())):
        current_node=q.get()
        if current_node!=None:
            print(current_node.data,end=' ')
        if current_node==None:
            if q.empty():
                break
            print()
            q.put(None)
            continue
        if current_node.left!=None:
            q.put(current_node.left)
        if current_node.right!=None:
            q.put(current_node.right)
        

#-------Taking input but level wise nodes-------#
def takingLevelWiseTreeInput():
    q=queue.Queue()
    print("Enter Root:")
    rootData=int(input())
    if rootData==-1:
        return None
    root=BinaryTreeNode(rootData)
    q.put(root)
    while (not(q.empty())):
        current_node=q.get()
        
        print("Enter left Child of ",current_node.data)
        leftChildData=int(input())
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            current_node.left=leftChild
            q.put(leftChild)
        print("Enter right Child of ",current_node.data)
        rightChildData=int(input())
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            current_node.right=rightChild
            q.put(rightChild)
    return root
root= takingLevelWiseTreeInput()
print()
printTreeDetailed(root)
print()
```

# Path Sum Root to Leaf


```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def rootToLeafPathsSumToKHelper(root, k, path, currSum) :
    if root is None :
        return

    if (root.left is None) and (root.right is None) :
        currSum += root.data
    if currSum == k :
        print(str(path + str(root.data) + " ").lstrip())
        return

    rootToLeafPathsSumToKHelper(root.left, k, str(path + str(root.data) + " "), (currSum +
                                                                                 root.data))
    rootToLeafPathsSumToKHelper(root.right, k, str(path + str(root.data) + " "), (currSum +
                                                                                  root.data))

def rootToLeafPathsSumToK(root, k):
    rootToLeafPathsSumToKHelper(root, k, "", 0)

#-------Taking input but level wise nodes-------#
def takingLevelWiseTreeInput():
    q=queue.Queue()
    print("Enter Root:")
    rootData=int(input())
    if rootData==-1:
        return None
    root=BinaryTreeNode(rootData)
    q.put(root)
    while (not(q.empty())):
        current_node=q.get()
        print("Enter left Child of ",current_node.data)
        leftChildData=int(input())
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            current_node.left=leftChild
            q.put(leftChild)
        print("Enter right Child of ",current_node.data)
        rightChildData=int(input())
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            current_node.right=rightChild
            q.put(rightChild)
    return root
root= takingLevelWiseTreeInput()
print()
node=int(input())
k=int(input())
rootToLeafPathsSumToK(root, k)
print()
```

# Print nodes at distance k from node


```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def printDK(root,k):
    if root == None:
        return 
    if k==0:
        print(root.data)
        return 
    printDK(root.left,k-1)
    printDK(root.right,k-1)
    return 
def nodesAtDistanceK(root, node, k) :
    if root is None:
        return -1
    if root.data == node:
        printDK(root,k)
        return 0
    leftchild=nodesAtDistanceK(root.left,node,k)
    if leftchild!=-1:
        if leftchild+1==k:
            print(root.data)
        else:
            printDK(root.right,k-leftchild-2)
        return 1+leftchild
    rightchild=nodesAtDistanceK(root.right,node,k)
    if rightchild!=-1:
        if rightchild+1==k:
            print(root.data)
        else:
            printDK(root.right,k-rightchild-2)
        return 1+rightchild
    return -1
            
    
            
                                                                    
#-------Taking input but level wise nodes-------#
def takingLevelWiseTreeInput():
    q=queue.Queue()
    print("Enter Root:")
    rootData=int(input())
    if rootData==-1:
        return None
    root=BinaryTreeNode(rootData)
    q.put(root)
    while (not(q.empty())):
        current_node=q.get()
        print("Enter left Child of ",current_node.data)
        leftChildData=int(input())
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            current_node.left=leftChild
            q.put(leftChild)
        print("Enter right Child of ",current_node.data)
        rightChildData=int(input())
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            current_node.right=rightChild
            q.put(rightChild)
    return root
root= takingLevelWiseTreeInput()
print()
node=int(input())
k=int(input())
nodesAtDistanceK(root, node, k)
print()
```
