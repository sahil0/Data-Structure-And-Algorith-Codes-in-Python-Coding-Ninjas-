# Preorder ,Postorder and Inorder Traversal :

```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
#---------------Preorder-------------------#
def printPreorder(root):
    if root==None:
        return 
    print(root.data,end=' ')
    printPreorder(root.left)        
    printPreorder(root.right)
#---------------Preorder-------------------#
def printPostorder(root):
    if root==None:
        return 
    printPostorder(root.left)        
    printPostorder(root.right)
    print(root.data,end=' ')
#---------------Preorder-------------------#
def printInorder(root):
    if root==None:
        return 
    printInorder(root.left) 
    print(root.data,end=' ')
    printInorder(root.right)
    
#-------Taking input nodes-------#
def treeInput():
    rootData=int(input())
    if rootData==-1:
        return None
    root =BinaryTreeNode(rootData)
    leftTree=treeInput()
    rightTree=treeInput()
    root.left=leftTree
    root.right=rightTree
    return root

root = treeInput()
print('PreorderSequence: ', end=' ')
printPreorder(root)
print()
print('PostOrderSequence: ',end=' ')
printPostorder(root)
print()
print('Inorder: ',end=' ')
printInorder(root)

#1
#2
#4
#-1
#-1
#5
#-1
#-1
#3
#-1
#7
#-1
#-1
#PreorderSequence:  1 2 4 5 3 
#PostOrderSequence:  4 5 2 3 1 
#Inorder:  4 2 5 1 3 
```

# Height of Tree

# Number of Leaf Nodes

# Print Nodes At Depth K

# Replace Node With Depth

# Nodes without sibling

```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
#---------------Printing nodes without sibling first left subtree and then of right subtree-------------------#
def printNodesWithoutSibling(root) :
    if root == None:
        return
    if root!=None and (root.left!=None or root.right!=None):
        if root.left==None and root.right!=None:
            print(root.right.data,end=' ')
        if root.right==None and root.left!=None:
            print(root.left.data,end=' ')
    printNodesWithoutSibling(root.left)
    printNodesWithoutSibling(root.right)
      
#-------Taking input nodes-------#
def treeInput():
    rootData=int(input())
    if rootData==-1:
        return None
    root =BinaryTreeNode(rootData)
    leftTree=treeInput()
    rightTree=treeInput()
    root.left=leftTree
    root.right=rightTree
    return root

root= treeInput()
print()
print("Nodes without siblings are:",end=' ')
printNodesWithoutSibling(root) 
```
